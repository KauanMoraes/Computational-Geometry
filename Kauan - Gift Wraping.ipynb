```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def casco_gift(lpontos):
    lpontos.sort(key=lambda x:x[1])
    lpontos=list(np.array(lpontos))
    p0=lpontos.pop(0)
    penult=p0+np.array([-1,0])
    casco=[p0]
    while len(casco)==1 or np.all(casco[-1]!=casco[0]):
        if len(casco)==2: lpontos.insert(0,p0)
        angs=[]
        for ponto in lpontos:
            linha=ponto-casco[-1]             
            linha/=np.linalg.norm(linha)
            ult_linha=penult-casco[-1]
            ult_linha/=np.linalg.norm(ult_linha)            
            ang=np.arccos(np.dot(linha,ult_linha))
            angs.append(ang)
        ang_max=max(angs)
        ind=angs.index(ang_max)
        penult=casco[-1]
        casco.append(lpontos.pop(ind))
    return np.array(casco)
```

```python
lpontos=[[2,10],[3,3],[4.5,6],[2.7,7.3],[5,7],[6,11],
         [7,2],[8,5],[4,10]]
apontos=np.array(lpontos)
c=casco_gift(lpontos)
```

```python
plt.plot(c.T[0],c.T[1])
plt.scatter(apontos.T[0],apontos.T[1])
```
