```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def casco(lpontos):
    lpontos.sort()
    lpontos=list(np.array(lpontos))
    lcasco=lpontos[:3]
    for i,ponto in enumerate(lpontos[3:]):
        i+=3
        iposits=[]
        for j,v in enumerate(lcasco):
            vet=np.cross(ponto-v,lcasco[j-1]-v)
            if vet>=0:
                iposits.append(j)
        p0=iposits[0]
        for posit in iposits[:-1][::-1]: lcasco.pop(posit)
        lcasco.insert(p0,ponto)
    return np.array((lcasco)+[lcasco[0]])     
```

```python
lpontos=[[2,10],[3,3],[4.5,6],[2.7,7.3],[5,7],[6,11],
         [7,2],[8,5],[4,10]]
apontos=np.array(lpontos)
c=casco(lpontos)
```

```python
plt.scatter(apontos.T[0],apontos.T[1])
plt.plot(c.T[0],c.T[1])
```

```python
l0=[[2,10],[3,9.8],[3.1,9.7],[4.5,8.9]]
a0=np.array(l0)
c=casco(l0)
```
