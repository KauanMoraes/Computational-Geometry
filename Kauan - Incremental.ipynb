```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def casco(lpontos):
    lpontos.sort()
    lpontos=list(np.array(lpontos))
    lcasco=lpontos[:3]
    if np.cross(lcasco[0]-lcasco[-1],lcasco[1]-lcasco[0])<0: #se lcasco for horário
            lcasco=[lcasco[0]]+lcasco[1::][::-1] # vira antihorário
    for i,ponto in enumerate(lpontos[3:]):
        i+=3
        iposits=[]
        for j,p in enumerate(lcasco):
            vet=np.cross(ponto-lcasco[j-1],p-lcasco[j-1])
            if vet>=0:
                iposits.append(j)
        if 0 not in iposits:
            p0=iposits[0]
            for index in iposits[:-1][::-1]: lcasco.pop(index)
        elif len(iposits)==1:  # se iposits só tiver 0
            p0=len(lcasco)
        else: # se iposits for da forma [0,j,j+1,...,len(lcasco)-1] i.e. se cineçar em 0 e pular alguns pontos 
              #(não pode haver outros dps do 0 pois teríamos que removê-lo)
            p0=iposits[1]
            for index in iposits[1:][::-1]: lcasco.pop(index)
        lcasco.insert(p0,ponto)
    return np.array(lcasco+[lcasco[0]])     
```

```python
lpontos=[[0,10],[11.2,3],[4.5,6],[7.6,7.3],[5.1,7],[6,11],
         [7,2],[1.2,5],[4,10],[4,5],[3,8],[7.6,9.1]]
apontos=np.array(lpontos)
c=casco(lpontos)
```

```python
plt.scatter(apontos.T[0],apontos.T[1])
plt.plot(c.T[0],c.T[1])
```
